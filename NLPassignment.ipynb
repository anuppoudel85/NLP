```python
import requests
from bs4 import BeautifulSoup

# FUNCTION: Get title and content of the given URL
def get_content_html(html):
  # Send a GET request to fetch the HTML content
  response = requests.get(html)
  content=""
  title=""
  # Check if the request was successful
  if response.status_code == 200:
      # Parse the HTML content
      soup = BeautifulSoup(response.content, 'html.parser')

      # Extract the content from the specified tags
      for tag in soup.find_all(['p', 'h2', 'h3', 'h4', 'h5']):
          content += tag.get_text() + "\n"
      for tag in soup.find_all(['h1']):
          title = tag.get_text()

  return content, title
```

```python
from pathlib import Path

# Make sure we have "data" directory
data_path = Path("data")
if not data_path.exists():
    Path.mkdir(data_path)
```

```python
htmls = ["https://www.discovertasmania.com.au/experiences/stories/top-10-attractions/",
        "https://www.taswalkingco.com.au/bruny-island-long-weekend/",
        "https://www.australia.com/en/trips-and-itineraries/best-road-trips-australia.html",
        "https://www.australia.com/en/trips-and-itineraries/best-wildflower-itineraries.html#field-of-dreams",
        "https://www.australia.com/en/places/hobart-and-surrounds/destinations-around-hobart.html",
        "https://www.australia.com/en/places/hobart-and-surrounds/best-restaurants.html",
        "https://www.australia.com/en/places/hobart-and-surrounds/guide-to-tasmanian-walking.html",
        "https://www.australia.com/en/trips-and-itineraries/hobart-and-surrounds/6-day-journey-through-tasmanias-western-wilds.html",
        "https://www.australia.com/en/things-to-do/walks-and-hikes/great-walks.html#three-capes-lodge-walk",
        "https://coalmines.org.au/history/",
        "https://portarthur.org.au/history/",
        "https://www.australiantraveller.com/tas/tasmanian-experiences/"]

doc_titles=[] # all extracted titles

# Save extracted content to local directories
for html in htmls:
  content, title = get_content_html(html)
  # the full text may reach the token limit per minutes for Groq API so we only take the part of the full text.
  content = content[:2000]
  doc_titles.append(title)
  with open(data_path / f"{title}.txt", "w") as fp:
    fp.write(content)
```

```python
# Load all documents
documents = [] # all documents used as knowledge base
for title in doc_titles:
    docs = SimpleDirectoryReader(
        input_files=[f"data/{title}.txt"]
    ).load_data()
    docs[0].doc_id = title
    documents.extend(docs)
```

## 2. Build Document Summary Index

```python
# default mode of building the index

splitter = SentenceSplitter(chunk_size=1024)

response_synthesizer = get_response_synthesizer(
    response_mode="tree_summarize", use_async=True
)
doc_summary_index = DocumentSummaryIndex.from_documents(
    documents,
    transformations=[splitter],
    response_synthesizer=response_synthesizer,
    show_progress=True,
)
```

## 3. Perform Retrieval from Document Summary Index

```python
from llama_index.core.indices.document_summary import (
    DocumentSummaryIndexEmbeddingRetriever,
)

retriever = DocumentSummaryIndexEmbeddingRetriever(
    doc_summary_index,
    similarity_top_k=3,
)
```

```python
# use retriever as part of a query engine
from llama_index.core.query_engine import RetrieverQueryEngine

# configure response synthesizer
response_synthesizer = get_response_synthesizer(response_mode="tree_summarize")

# assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    response_synthesizer=response_synthesizer,
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Example: User reviews (Positive = 1, Negative = 0)
reviews = [
    "The response was very helpful and accurate", 1,
    "I didn't find the information useful", 0,
    "The summary was concise and well-written", 1,
    "The response was irrelevant to my query", 0
]
texts, labels = reviews[::2], reviews[1::2]

# Convert text to numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.25, random_state=42)

# Train Naive Bayes classifier
nb = MultinomialNB()
nb.fit(X_train, y_train)

# Evaluate performance
y_pred = nb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import os
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
from rouge import Rouge
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    load_index_from_storage,
    StorageContext,
    Settings,
)
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core import get_response_synthesizer
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.groq import Groq

# Set up Groq API key
os.environ["GROQ_API_KEY"] = "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"

# Set up a local embedding model
local_embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Set up LLaMA model via Groq
llm_70b = Groq(model="llama3-70b-8192")

# Update the global settings
Settings.embed_model = local_embed_model
Settings.llm = llm_70b

# Load documents and create index
documents = SimpleDirectoryReader('data').load_data()
index = VectorStoreIndex.from_documents(documents)
index.storage_context.persist('vector_index')

# Load the index
storage_context = StorageContext.from_defaults(persist_dir="vector_index")
loaded_index = load_index_from_storage(storage_context)

# Initialize retriever
retriever = VectorIndexRetriever(
    index=loaded_index,
    similarity_top_k=5,
)

# Configure response synthesizer
response_synthesizer = get_response_synthesizer(response_mode="tree_summarize")

# Assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    response_synthesizer=response_synthesizer,
)

# Define ground truth data
ground_truth = {
    "What are the top attractions in Tasmania?": "The top attractions include Cradle Mountain, MONA, and Freycinet.",
    "What can I do on Bruny Island?": "You can enjoy guided hikes, gourmet food, and scenic views.",
    "Is Port Arthur open all year round?": "Yes, the site is open year-round with daily tours available.",
    "What activities are available in Tasmania in winter?": "Winter activities include hiking, spa retreats, and food festivals.",
    "Can I swim at Wineglass Bay?": "Yes, Wineglass Bay offers swimming, though the water can be cold.",
    "How can I travel between Hobart and Bruny Island?": "You can take a ferry from Kettering to Bruny Island.",
    "Is camping allowed at Cradle Mountain?": "Yes, there are designated camping areas in Cradle Mountain.",
    "What makes MONA unique?": "MONA offers a blend of modern art, architecture, and unique exhibitions.",
    "Are guided tours available at Port Arthur?": "Yes, guided tours are available daily at the historic site."
}

# Evaluation metrics
def calculate_bleu(reference, candidate):
    return sentence_bleu([reference.split()], candidate.split(), smoothing_function=SmoothingFunction().method1)

def calculate_rouge(reference, candidate):
    rouge = Rouge()
    scores = rouge.get_scores(candidate, reference)
    return scores[0]['rouge-l']['f']

def evaluate_retrieval(query, retriever, expected_answer):
    retrieved_nodes = retriever.retrieve(query)
    retrieved_text = " ".join([node.node.text for node in retrieved_nodes])
    keywords = expected_answer.lower().split()
    return any(keyword in retrieved_text.lower() for keyword in keywords)

# Perform evaluation
bleu_scores = []
rouge_scores = []
retrieval_scores = []

for question, true_answer in ground_truth.items():
    # Get system's answer
    system_answer = query_engine.query(question).response

    # Calculate BLEU score
    bleu = calculate_bleu(true_answer, system_answer)
    bleu_scores.append(bleu)

    # Calculate ROUGE score
    rouge = calculate_rouge(true_answer, system_answer)
    rouge_scores.append(rouge)

    # Evaluate retrieval
    retrieval_success = evaluate_retrieval(question, retriever, true_answer)
    retrieval_scores.append(1 if retrieval_success else 0)

    print(f"Question: {question}")
    print(f"True Answer: {true_answer}")
    print(f"System Answer: {system_answer}")
    print(f"BLEU Score: {bleu:.4f}")
    print(f"ROUGE Score: {rouge:.4f}")
    print(f"Retrieval Success: {retrieval_success}")

    # Print retrieved documents for analysis
    #retrieved_nodes = retriever.retrieve(question)
   # print("\nRetrieved Documents:")
    #for i, node in enumerate(retrieved_nodes, 1):
        #print(f"Document {i}: {node.node.text[:200]}...")  # Print first 200 characters

    #print("-" * 50)

# Calculate average scores
avg_bleu = np.mean(bleu_scores)
avg_rouge = np.mean(rouge_scores)
retrieval_accuracy = np.mean(retrieval_scores)

print(f"Average BLEU Score: {avg_bleu:.4f}")
print(f"Average ROUGE Score: {avg_rouge:.4f}")
print(f"Retrieval Accuracy: {retrieval_accuracy:.4f}")
```
